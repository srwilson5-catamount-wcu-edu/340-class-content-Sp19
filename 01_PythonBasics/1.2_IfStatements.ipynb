# Relational Operators

| Symbol | Task Performed |
|----|---|
| == | True, if it is equal |
| !=  | True, if not equal to |
| < | less than |
| > | greater than |
| <=  | less than or equal to |
| >=  | greater than or equal to |

# If Statement 
**Enter the following code in the cell below and run:**  
```Python
numberToCheck = 15

# Check if the number is greater than 10
if numberToCheck > 10:
    print("The number is greater than 10")
else:
    print("That number is not greater than 10!")
```

````python
numberToCheck = -15

# Check if the number is greater than 10
if numberToCheck > 10:
    print("The number is greater than 10")
else:
    print("That number is not greater than 10!")
````

# Elif and Else

**Enter the following code in the cell below and run:**  
```Python
exam_score = 33

if exam_score >= 90:
    exam_grade = "A"
elif exam_score >= 80:
    exam_grade = "B"
else:
    exam_grade = "unknown"

print("Your Exam Grade is:", exam_grade)
```

````python
exam_score = 33

if exam_score >= 90:
    exam_grade = "A"
elif exam_score >= 80:
    exam_grade = "B"
else:
    exam_grade = "unknown"

print("Your Exam Grade is:", exam_grade)
````
